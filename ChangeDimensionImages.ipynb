```python
from PIL import Image
from IPython.display import Image as IPImage, display
```

```python
def resize_image(input_path, output_path, new_width, new_height):
    # Open the image
    image = Image.open(input_path)

    # Resize the image
    resized_image = image.resize((new_width, new_height), Image.ANTIALIAS)

    # Save the resized image
    resized_image.save(output_path)

    print("Image resized successfully!")

    # Display the resized image in the notebook
    display(IPImage(output_path))

if __name__ == "__main__":
    input_path = "gray_profile.jpg"  # Replace with the actual input image path
    output_path = "gray_profile_resized.jpg"    # Replace with the desired output path
    new_width = 6200
    new_height = 3600

    resize_image(input_path, output_path, new_width, new_height)
```
